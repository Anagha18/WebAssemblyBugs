```python
import requests

# Set your search parameters
owner = "torch2424"
repo = "wasm-by-example"
token = ""  # Replace with personal token
keyword = "wasm"  # Replace with desired keyword
label = "bug"      # Label to filter issues (optional)

# Construct the search query string
query = f"repo:{owner}/{repo} {keyword} label:{label} state:closed"

# Construct the URL
url = "https://api.github.com/search/issues"
headers = {
    "Authorization": f"token {token}",
    "Accept": "application/vnd.github.v3+json"
}

# Set up parameters for the request
params = {
    "q": query
}

# Make the request
response = requests.get(url, headers=headers, params=params)
# Check if the response was successful
if response.status_code == 200:
    issues = response.json().get("items", [])
    for issue in issues:
        print(f"Issue Title: {issue['title']}")
        print(f"URL: {issue['html_url']}")
        print(f"Snippet: {issue['body']}...")
        print()
else:
    print(f"Request failed with status code {response.status_code}")
    print("Response:", response.json())
```
